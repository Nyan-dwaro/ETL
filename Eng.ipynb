```python
#ETRACT
#funcion to give a bit of info
def extract(file_path):
    data = pd.read_csv(file_path)
    #some basic info
    print(f"A bit of information about the data in {file_path}:")
    print(f"There are {data.shape[0]} rows and {data.shape[1]} columns in the dataset")
    print("The datatypes are as follows:")
    print(data.dtypes)

    return data

#call the function
app_data = extract("/content/1.csv")
review_data = extract("/content/2.csv")
app_data.head()
```

```python
#Transformation
def transform(apps, reviews, category, min_rating, min_reviews):
  print(f"Transforming data to curate a dataset with all {category} apps and their corresponding"
  f"reviews with a rating of atleast {min_rating}"
  f"and {min_reviews} ratings")

  #drop duplicates
  reviews = reviews.drop_duplicates()
  apps = apps.drop_duplicates(['App'])

  # find all the apps and reviews in the food and drinks category
  subset_apps = apps.loc[apps['Category'] == category, :]
  subset_reviews = reviews.loc[reviews["App"].isin(subset_apps["App"]), ["App", "Sentiment_Polarity"]]

  #aggregated subset_review dataframe
  aggregated_review = subset_reviews.groupby("App").mean()

  #join them back
  joined_apps_reviews = subset_apps.join(aggregated_review, on = "App", how = "left")

  #keep only the needed columns
  filtered_apps_reviews = joined_apps_reviews.loc[:, ['App', 'Rating', 'Reviews', 'Sentiment_Polarity', 'Installs']]

  # convert reviews to int
  filtered_apps_reviews = filtered_apps_reviews.astype({"Reviews": "int32"})

  #Return ratings greater than 4 and reviews more than 1000
  top_apps = filtered_apps_reviews.loc[(filtered_apps_reviews["Rating"]>min_rating) & (filtered_apps_reviews["Reviews"]>min_reviews), : ]

  top_apps.sort_values(by = ["Rating", "Reviews"], ascending = False, inplace = True)
  top_apps.reset_index(drop = True, inplace = True)


  return top_apps

#function call
top_apps_data = transform(
    apps = app_data,
    reviews = review_data,
    category = "FOOD_AND_DRINK",
    min_rating = 4.0,
    min_reviews = 1000
)
print(top_apps_data)
```
